```python
import re

name = 'John Doe'

resObj = re.search(r"^([a-zA-Z]{2,}\s[a-zA-Z]{1,}'?-?[a-zA-Z]{2,}\s?([a-zA-Z]{1,})?)", name)
if resObj:
    res = resObj.group(1)
else:
    raise ValueError(f"Invalid account name. Account name must be first name followed by last name.")


res
```

```python
def ca_withdraw(amount, opening_balance=10, overdraft_limit=50):
    print(f"Opening balance: {opening_balance}")
    if opening_balance + amount > overdraft_limit:
        print("You can't go over your agreed overdraft limit!")
    opening_balance -= amount
    # print(f"Closing balance: {opening_balance}")

ca_withdraw(amount=45, opening_balance=10, overdraft_limit=50)
```
